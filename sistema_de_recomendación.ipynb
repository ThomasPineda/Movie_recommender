*   Sistemas basados en popularidad
*   Sistemas basados en contenido
*   Sistemas colaborativos

  1.1.1 Sistemas basados en popularidad

  Estos sistemas generalmente son usados como sistema de recomendación inicial en cualquier plataforma de oferta de contenido; ya que inicialmente se carece de datos del usuario y de historiales de consumo entonces se hace imposible personalizar el contenido sugerido. Estos sistemas basados en popularidad son los menos complicados de implementar ya que no segmentan el publico objetivo, sin embargo son los menos personalizados y los que menos prcisión presentan.

  1.1.2. Sistemas basados en contenido

  Los sistemas basados en contenido son los sistemas más utilizados en la actualidad, ya que han demostrado una alta efectividad en este mundo de los sistemas de recomendación.  Estos sistemas se basan principalmente en el historial de consumo de un usuario puntual y deja de lado caracteristicas propias del individuo.  Se enfoca en el contenido consumido con anterioridad y recomienda contenido relacionado con este historial.  El desarrollo de estos sistemas apunta a relacionar contenidos entre si aprovechandose de la invarianza de los gustos generales de las personas y esforzandose por agrupar contenido que satisfaga un perfil de gustos.

  1.1.3 Sistemas colaborativos

  Este tipo de sistemas apunta a individualizar perfiles conociendo los datos del usuario.  Intenta ubicar al usuario en un grupo con perfiles sismilares para recomendar contenido ya consumido por los integrantes de su mismo grupo.  Estos sistemas como se puede inferir, no se centran en el contenido sino en el usurio.

5.1 **Sistema basado en contenidos**

  Los sistemas de recomendación basados en contenido son un tipo de sistema de recomendación que se basa en las características y el contenido de los elementos que se están recomendando. Estos sistemas utilizan información detallada sobre los elementos y las preferencias del usuario para realizar recomendaciones personalizadas.

  **Ventajas de los Sistemas de Recomendación Basados en Contenido:**

  **Explicabilidad:** Los sistemas de recomendación basados en contenido son transparentes y explicables, ya que las recomendaciones se generan en función de las características de los elementos y las preferencias del usuario.

  **Fácil implementación inicial:** Son relativamente fáciles de implementar, especialmente cuando se dispone de información detallada sobre los elementos.

  **Independencia de la comunidad:** No dependen de las calificaciones de otros usuarios y, por lo tanto, son útiles cuando los datos de calificación de usuarios son escasos.

  **Limitaciones de los Sistemas de Recomendación Basados en Contenido:**

  **Falta de serendipia:** Pueden carecer de la capacidad de descubrir elementos inesperados o sorprendentes, ya que se basan en la concordancia de características conocidas.

  **Sobreespecialización:** Pueden tender a recomendar elementos muy similares a los que el usuario ya ha consumido, lo que limita la diversidad de las recomendaciones.

  **Necesidad de datos detallados:** Requieren información detallada sobre las características de los elementos, lo que a veces no está disponible.

  En resumen, los sistemas de recomendación basados en contenido son una categoría importante de sistemas de recomendación que utilizan información detallada sobre los elementos y el perfil del usuario para proporcionar recomendaciones personalizadas. Son especialmente útiles cuando se dispone de datos detallados de los elementos, pero también presentan algunas limitaciones en términos de diversidad de recomendaciones.

*TF-IDF*

El TF-IDF Term Frequency - Inverse Document Frequency. Esta expresión puede ser traducida para el español como "Frecuencia del Término - Frecuencia Inversa de los Documentos". tiene en cuenta el número de veces que aparece la palabra (o token) en dicho documento, pero también el total de veces que aparece en todo el corpus.

El algoritmo de TF-IDF lo que hace es asignar un peso respecto a lo relevante que es una palabra en un documento. De este modo, los tokens que son muy frecuentes a nivel de documento y de corpus (por ejemplo, las stopwords) van a tener un TF-IDF muy bajo.


image.png


En resumen, esta matriz TF-IDF asigna un número a cada palabra en las descripciones de las películas para indicar cuán importantes son esas palabras para cada película. Ayuda a la computadora a entender y comparar las películas en función de las palabras que utilizan en sus descripciones.

```python
#Calcular el TF-IDF de la descripción de las películas y hacer una matriz
tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
#Convierte los calculos en una matriz, en este caso cada fila es cada película y las columnas el TF-IDF de las palabras que aparecen en las descricpciones.
tfidf_matrix = tf.fit_transform(smd['description'])
tfidf_matrix
```

*Similitud del coseno*

Para saber la cercanía entre películas se procede con esta metrica que lo que hace es calcular la similitud entre dos o más vectores. La similitud del coseno es el coseno del ángulo entre vectores. Los vectores suelen ser distintos a cero y están dentro de un espacio de producto escalar.

La similitud del coseno en el aprendizaje automático se utiliza para tareas de clasificación, como una métrica en los algoritmos de clasificación KNN para encontrar el número óptimo de vecinos y también en el modelo KNN. En los sistemas de recomendación, se utiliza con el mismo principio de ángulos de coseno, donde incluso si la similitud del contenido es menos similar, se consideraría como el contenido menos recomendado, y para mayor similitud de contenidos, las recomendaciones generadas estarían en la parte superior. La similitud del coseno también se utiliza en datos textuales para encontrar la similitud entre los textos vectorizados a partir del documento de texto original.

linear_kernel es una función que calcula la similitud del coseno entre vectores. La similitud del coseno es una medida de cuán similares son dos vectores. En este caso, los vectores representan las descripciones de películas.

Al aplicar linear_kernel(tfidf_matrix), se calcula la similitud del coseno entre cada par de filas en tfidf_matrix. En otras palabras, se compara cada película con todas las demás películas para determinar cuán similares son en función de sus descripciones. El resultado es una matriz cuadrada donde cada entrada (i, j) representa la similitud del coseno entre la película i y la película j.

```python
#Calcula la similitud del coseno entre cada vector de la matriz
cosine_sim = linear_kernel(tfidf_matrix)
```

```python
#Realiza la recomendación de las películas de más similares a menos similares
def get_recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:31]
    movie_indices = [i[0] for i in sim_scores]
    return titles.iloc[movie_indices]
```

5.2 **Sistema de recomendación por filtro colaborativo**

Los sistemas de recomendación por filtro colaborativo son una categoría importante de sistemas de recomendación que se basan en la idea de que las preferencias y gustos de un usuario pueden ser predictivos de las preferencias de otros usuarios con gustos similares. Estos sistemas utilizan la colaboración o interacción de un conjunto de usuarios para hacer recomendaciones a un usuario en particular. Aquí se explica de manera detallada cómo funcionan los sistemas de recomendación por filtro colaborativo:

**1. Recopilación de Datos:**

El primer paso en la construcción de un sistema de recomendación por filtro colaborativo es recopilar datos de usuarios y elementos. Estos datos pueden incluir calificaciones, reseñas, compras, clics, historiales de navegación u otras interacciones entre usuarios y elementos. La información debe ser recopilada en una matriz donde las filas representan usuarios y las columnas representan elementos.

**2. Matriz de Utilidad:**

A partir de los datos recopilados, se crea una matriz de utilidad (también conocida como matriz de calificaciones), que muestra cómo los usuarios han valorado o interactuado con los elementos. En esta matriz, las celdas contienen calificaciones numéricas, indicando la opinión de un usuario sobre un elemento o, en el caso de sistemas binarios, si un usuario interactuó o no con un elemento.

**3. Identificación de Usuarios Similares:**

El sistema busca usuarios que sean similares al usuario para el que se están haciendo recomendaciones. La similitud se puede medir de varias maneras, pero una métrica común es la similitud coseno. Cuanto más similar sea el perfil de dos usuarios en términos de sus calificaciones o interacciones pasadas, mayor será su similitud.

**4. Filtrado de Usuarios Similares:**

Una vez identificados los usuarios similares, el sistema selecciona un subconjunto de ellos para la recomendación. Este subconjunto puede ser de tamaño fijo o puede variar según el contexto. Los usuarios similares tienen un peso mayor en la recomendación que aquellos que son menos similares.

**5. Predicción de Calificaciones:**

El sistema calcula las calificaciones que el usuario en cuestión asignaría a los elementos que no ha visto o interactuado aún. Esto se hace considerando las calificaciones de los usuarios similares para esos elementos y aplicando técnicas de interpolación o regresión. Una forma común de hacerlo es calcular el promedio ponderado de las calificaciones de los usuarios similares para cada elemento no calificado.

**6. Generación de Recomendaciones:**

El sistema genera una lista de elementos recomendados para el usuario en función de las calificaciones predichas. Los elementos se ordenan de manera que los que tienen las calificaciones más altas sean recomendados primero.

**7. Realimentación y Aprendizaje Continuo:**

A medida que el usuario interactúa con las recomendaciones y proporciona retroalimentación (por ejemplo, calificando elementos recomendados), el sistema de recomendación puede mejorar sus predicciones y recomendaciones futuras.

Existen dos enfoques principales de filtro colaborativo: el filtrado colaborativo basado en usuario y el filtrado colaborativo basado en elementos.

- Filtrado Colaborativo Basado en Usuario: Este enfoque compara a un usuario con otros usuarios similares para hacer recomendaciones. Se basa en la idea de que si dos usuarios tienen gustos similares, es probable que compartan preferencias en otros elementos.

- Filtrado Colaborativo Basado en Elementos: En lugar de comparar usuarios, este enfoque se basa en la similitud entre los elementos en sí. Si a un usuario le gustó un conjunto de elementos, se recomendarán elementos similares a los que ya le gustaron.

En resumen, los sistemas de recomendación por filtro colaborativo se basan en la colaboración y la similitud entre usuarios para proporcionar recomendaciones personalizadas. Estos sistemas son ampliamente utilizados en plataformas de comercio electrónico, servicios de streaming de contenido, redes sociales y muchas otras aplicaciones donde la personalización es importante.

MSE básicamente mide el error cuadrado promedio de nuestras predicciones. Para cada punto, calcula la diferencia cuadrada entre las predicciones y el objetivo y luego promedia esos valores.

Cuanto mayor sea este valor, peor es el modelo. Nunca es negativo, ya que estamos cuadrando los errores de predicción individuales antes de sumarlos, pero sería cero para un modelo perfecto.

```python
#Obtener MSE de las predicciones
def get_mse(preds, actuals):
    if preds.shape[1] != actuals.shape[1]:
        actuals = actuals.T
    preds = preds[actuals.nonzero()].flatten()
    actuals = actuals[actuals.nonzero()].flatten()
    return mean_squared_error(preds, actuals)

# Realizo las predicciones para el test set
users_predictions_test = sim_matrix.dot(ratings) / np.array([np.abs(sim_matrix).sum(axis=1)]).T
users_predictions_test = users_predictions_test[536:671,:]
get_mse(users_predictions, ratings_train)
```
